```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Directorio de la carpeta con los archivos de audio
carpeta_origen = '/content/drive/MyDrive/Procesamiento de datos secuenciales/Audios/Audios'

# Directorio de la carpeta donde se guardarán los espectrogramas
carpeta_destino = '/content/drive/MyDrive/Procesamiento de datos secuenciales/Audios/Espect'

# Obtener la lista de archivos de audio en la carpeta de origen
archivos_audio = os.listdir(carpeta_origen)

# Recorrer cada archivo de audio
for archivo in archivos_audio:
    # Ruta completa del archivo de audio
    ruta_audio = os.path.join(carpeta_origen, archivo)
    
    # Cargar el archivo de audio utilizando librosa
    audio, sr = librosa.load(ruta_audio)
    
    # Generar el espectrograma utilizando librosa
    espectrograma = librosa.feature.melspectrogram(y=audio, sr=sr)
    
    # Mostrar el espectrograma utilizando matplotlib
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(librosa.power_to_db(espectrograma, ref=np.max), y_axis='mel', x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Espectrograma')
    plt.tight_layout()
    
    # Guardar el espectrograma como imagen en la carpeta de destino
    nombre_espectrograma = archivo.replace('.wav', '.png')
    ruta_espectrograma = os.path.join(carpeta_destino, nombre_espectrograma)
    plt.savefig(ruta_espectrograma)
    
    # Cerrar la figura para liberar memoria
    plt.close()
```

```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Directorio de la carpeta con los archivos de audio
carpeta_origen = '/content/drive/MyDrive/Procesamiento de datos secuenciales/Audios/Audios'

# Directorio de la carpeta donde se guardarán los espectrogramas
carpeta_destino = '/content/drive/MyDrive/Procesamiento de datos secuenciales/Audios/Espectograma'

# Obtener la lista de archivos de audio en la carpeta de origen
archivos_audio = os.listdir(carpeta_origen)

# Recorrer cada archivo de audio
for archivo in archivos_audio:
    # Ruta completa del archivo de audio
    ruta_audio = os.path.join(carpeta_origen, archivo)
    
    # Cargar el archivo de audio utilizando librosa
    audio, sr = librosa.load(ruta_audio)
    
    # Generar el espectrograma utilizando librosa
    espectrograma = librosa.feature.melspectrogram(y=audio, sr=sr)
    
    # Convertir el espectrograma a decibeles utilizando librosa
    espectrograma_db = librosa.power_to_db(espectrograma, ref=np.max)
    
    # Guardar el espectrograma como imagen en la carpeta de destino
    nombre_espectrograma = archivo.replace('.wav', '.png')
    ruta_espectrograma = os.path.join(carpeta_destino, nombre_espectrograma)
    librosa.display.specshow(espectrograma_db, sr=sr, x_axis='time', y_axis='mel')
    plt.axis('off')  # Para eliminar los ejes
    plt.savefig(ruta_espectrograma, bbox_inches='tight', pad_inches=0)
    plt.close()
```
