```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Ruta de la carpeta de entrada y salida
carpeta_entrada = "/content/drive/MyDrive/Procesamiento de datos secuenciales/Audios/Audios"
carpeta_salida = "/content/drive/MyDrive/Procesamiento de datos secuenciales/Audios/MFCC2"

# Obtener la lista de archivos de audio en la carpeta de entrada
archivos_audio = [archivo for archivo in os.listdir(carpeta_entrada) if archivo.endswith(".wav")]

# Recorrer cada archivo de audio
for archivo_audio in archivos_audio:
    # Cargar el archivo de audio
    ruta_audio = os.path.join(carpeta_entrada, archivo_audio)
    audio, sr = librosa.load(ruta_audio, sr=None)
    
    # Extraer los coeficientes MFCC del audio
    mfcc = librosa.feature.mfcc(audio, sr=sr)
    
    # Normalizar los coeficientes MFCC
    mfcc_norm = (mfcc - np.mean(mfcc)) / np.std(mfcc)
    
    # Generar la visualizaci贸n de los coeficientes MFCC
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mfcc_norm, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.tight_layout()
    
    # Generar el nombre del archivo de salida
    nombre_archivo_salida = archivo_audio.replace(".wav", ".png")
    ruta_salida = os.path.join(carpeta_salida, nombre_archivo_salida)
    
    # Guardar la visualizaci贸n de los coeficientes MFCC como imagen PNG
    plt.savefig(ruta_salida, format='png')
    plt.close()
```

```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Ruta de la carpeta de entrada y salida
carpeta_entrada = "/content/drive/MyDrive/Procesamiento de datos secuenciales/Audios/Audios"
carpeta_salida = "/content/drive/MyDrive/Procesamiento de datos secuenciales/Audios/MFCC2"


# Obtener la lista de archivos de audio en la carpeta de entrada
archivos_audio = [archivo for archivo in os.listdir(carpeta_entrada) if archivo.endswith(".wav")]

# Recorrer cada archivo de audio
for archivo_audio in archivos_audio:
    # Cargar el archivo de audio
    ruta_audio = os.path.join(carpeta_entrada, archivo_audio)
    audio, sr = librosa.load(ruta_audio, sr=None)
    
    # Extraer los coeficientes MFCC del audio
    mfcc = librosa.feature.mfcc(audio, sr=sr)
    
    # Normalizar los coeficientes MFCC
    mfcc_norm = (mfcc - np.mean(mfcc)) / np.std(mfcc)
    
    # Generar la visualizaci贸n de los coeficientes MFCC
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mfcc_norm, x_axis='time')
    plt.axis('off')  # Desactivar los ejes
    plt.tight_layout()
    
    # Generar el nombre del archivo de salida
    nombre_archivo_salida = archivo_audio.replace(".wav", ".png")
    ruta_salida = os.path.join(carpeta_salida, nombre_archivo_salida)
    
    # Guardar la visualizaci贸n de los coeficientes MFCC como imagen PNG
    plt.savefig(ruta_salida, format='png', bbox_inches='tight', pad_inches=0)
    plt.close()
```
